```python
#Controlling Axes and adding Diagonals
g = sns.FacetGrid(data=movies,row='Genre',col='Year',hue='Genre')
kws={
    #'edgecolor':'black',
    'linewidth':0.5,'s':50}
g = g.map(sns.scatterplot,'CriticRating','AudienceRating',**kws)
g.set(xlim=(0,100),ylim=(0,100))
for ax in g.axes.flat:
    #axes is a matrix of each plot in the Facet Grid
    #flat flattens the matrix to a 1D array
    ax.plot((0,100),(0,100),c='gray',ls='--')
g.add_legend()
```

```python
#Building Dashboards
sns.set_style('dark')
f, axes = plt.subplots(2,2,figsize=(10,10))

k1 = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,ax=axes[0,0])
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,ax=axes[0,1])

z = sns.violinplot(data=movies[movies.Genre=='Drama'], x='Year',y='CriticRating',ax=axes[1,0])

k4 = sns.kdeplot(movies.CriticRating,movies.AudienceRating,shade=True,shade_lowest=False,cmap='Reds',ax=axes[1,1])
#Tip: Smooth out the edges by overlaying the viz with lines instead of shade
k4b = sns.kdeplot(movies.CriticRating,movies.AudienceRating,shade_lowest=False,cmap='Reds',ax=axes[1,1])

#axes[1,1].hist(movies.CriticRating,bins=15) #Different because hist is a plt function, not sns
#also axes has visibility to hist, but not any sns functions

k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
#Styling Dashboards

sns.set_style('darkgrid',{'axes.facecolor':'black'})
f, axes = plt.subplots(2,2,figsize=(10,10))

#Plot [0,0]
k1 = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,shade=True
                 ,shade_lowest=True,cmap='inferno'
                 ,ax=axes[0,0])
k1b = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating
                 ,shade_lowest=True,cmap='cool'
                 ,ax=axes[0,0])

#Plot [0,1]
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,shade=True
                 ,shade_lowest=True,cmap='inferno'
                 ,ax=axes[0,1])
k2b = sns.kdeplot(movies.BudgetMillions,movies.CriticRating
                 ,shade_lowest=True,cmap='cool'
                 ,ax=axes[0,1])

#Plot [1,0]
z = sns.violinplot(data=movies
                   , x='Year',y='BudgetMillions'
                   #,shade_lowest=False,cmap='Reds'
                   ,palette='YlOrRd'
                   ,ax=axes[1,0])

#Plot [1,1]
k4 = sns.kdeplot(movies.CriticRating,movies.AudienceRating
                 ,shade=True,shade_lowest=False,cmap='Blues_r'
                 ,ax=axes[1,1])
#Tip: Smooth out the edges by overlaying the viz with lines instead of shade
k4b = sns.kdeplot(movies.CriticRating,movies.AudienceRating
                  ,shade_lowest=False,cmap='gist_gray_r'
                  ,ax=axes[1,1])

k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
#Thematic elements
list1 = list()
for genre in movies.Genre.cat.categories:
            list1.append(movies[movies.Genre == genre].BudgetMillions)

sns.set_style('whitegrid')
fig, ax = plt.subplots() #create a subplot of size (1,1) to gain more control of the size of the chart
fig.set_size_inches(11,8.5) #size of US letter paper
h = plt.hist(list1,bins=30,stacked=True,label=movies.Genre.cat.categories)
plt.title('Movie Budget Distribution',fontsize=30,fontstyle='italic',color='DarkBlue',fontname='Console')
plt.ylabel('Number of Movies',fontsize=14,color='DarkRed')
plt.xlabel('Budget',fontsize=14,color='DarkGreen')
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.legend(frameon=True,fancybox=True,shadow=True,framealpha=0.35,prop={'size':14})
plt.show()
```
